```python
import matplotlib.pyplot as plt
```

```python
least_popular = job.civil_service_title.value_counts()
```

```python
popular_titles = least_popular.iloc[:20]
```

```python
sns.catplot(kind='count',data=job,y='civil_service_title',order=popular_titles.index)
```

```python
salary_table = job[['civil_service_title','salary_range_low','salary_range_high']]
jobs_highest_low_range = pd.DataFrame(salary_table.groupby(['civil_service_title'])['salary_range_low'].mean().nlargest(10)).reset_index()
```
