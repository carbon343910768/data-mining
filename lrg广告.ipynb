```python
from sklearn.linear_model import LinearRegression
import sklearn.datasets as datasets
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
path='/home/carbon13/machine_learning_study_follow_yincheng-master/logist'
```

```python
data = pd.read_csv(path + '/Advertising.csv')
data.info()
```

```python
y=data.sales
```

```python
x=data[['TV','radio','newspaper']]
```

```python
plt.plot(data.TV,y,'ro',label='TV')
plt.plot(data.radio,y,'g^',label='radio')
plt.plot(data.newspaper,y,'mv',label='newspaper')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(9,10))
plt.subplot(311)
plt.plot(data.TV,y,'ro',label='TV')
plt.grid()
plt.legend()
plt.subplot(312)
plt.plot(data.radio,y,'g^',label='radio')
plt.grid()
plt.legend()
plt.subplot(313)
plt.plot(data.newspaper,y,'mv',label='newspaper')
plt.grid()
plt.legend()
plt.show()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y)
lrg = LinearRegression()
lrg.fit(x_train,y_train)
```

```python
lrg.coef_
```

```python
lrg.intercept_
```

```python
lrg.score(x_test,y_test)
```

```python
y_predict = lrg.predict(x_test)
y_predict.shape
```

```python
y_test.shape
```

```python
t=np.arange(len(x_test))
t.shape
plt.plot(t,y_test,'r',label='test')
plt.plot(t,y_predict,'g',label='predict')
plt.grid()
plt.legend()
plt.show()
```
